```python
# Imports
import torch
from abstractions3d.primitives.shapes import Box3D
from abstractions3d.primitives.visualization import visualize_boxes_3d
from abstractions3d.dsl.nodes import Cube, Move3D, Union3D, SymTrans3D, SymRef3D
```

```python
# Create some 3D shapes
cube1 = Cube(2, 1, 1)
cube2 = Move3D(cube1, t_x=3, t_y=0, t_z=1)
union = Union3D(cube1, cube2)
sym = SymTrans3D(union, axis="x", dist=6, degree=3)
sym_ref = SymRef3D(sym, axis="y")

# Convert the top-level shape to Box3D objects
boxes = sym_ref.get_box3d_list()

# Visualize
visualize_boxes_3d(boxes)
```

```python
# Chair dimensions
seat = Cube(2.0, 0.5, 2.0)  # width, height, depth
backrest = Move3D(Cube(2.0, 2.0, 0.5), t_x=0, t_y=1.25, t_z=-0.75)  # behind seat
leg = Cube(0.25, 1.0, 0.25)

# Legs positions
leg1 = Move3D(leg, t_x=-0.875, t_y=-0.25, t_z=-0.875)
leg2 = Move3D(leg, t_x=0.875, t_y=-0.25, t_z=-0.875)
leg3 = Move3D(leg, t_x=-0.875, t_y=-0.25, t_z=0.875)
leg4 = Move3D(leg, t_x=0.875, t_y=-0.25, t_z=0.875)

# Combine everything
legs_union = Union3D(Union3D(leg1, leg2), Union3D(leg3, leg4))
seat_and_legs = Union3D(seat, legs_union)
chair = Union3D(seat_and_legs, backrest)
```

```python
# Convert to Box3D objects
boxes = chair.get_box3d_list()

# Visualize
visualize_boxes_3d(boxes)
```
