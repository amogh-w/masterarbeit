```python
import sys
import os

# Add the src directory to sys.path
sys.path.append(os.path.abspath("../src"))
```

```python
# Imports
import torch
from abstractions3d.dsl.nodes import Rect3D, Move3D, Union3D, SymTrans3D, SymRef3D
from abstractions3d.primitives.visualization import visualize_boxes_3d
from abstractions3d.dsl.instantiation import instantiate_3d
```

```python
print("=== Basic Rect3D ===")
box = Rect3D(1.0, 1.0, 1.0)
print(box)
visualize_boxes_3d(box.get_box3d_list())
```

```python
# Move Operation
print("\n=== Move3D ===")
moved_box = Move3D(box, 2.0, 0.0, 0.0)
print(moved_box)
visualize_boxes_3d(moved_box.get_box3d_list())
```

```python
# Union Operation
print("\n=== Union3D ===")
box2 = Rect3D(0.5, 2.0, 0.5)
union = Union3D(box, box2)
print(union)
visualize_boxes_3d(union.get_box3d_list())
```

```python
# Symmetric Translation
print("\n=== SymTrans3D ===")
row_of_boxes = SymTrans3D(box, "x", 3.0, 4)
print(row_of_boxes)
visualize_boxes_3d(row_of_boxes.get_box3d_list())
```

```python
# Symmetric Reflection
print("\n=== SymRef3D ===")
mirrored = SymRef3D(box, "z")
print(mirrored)
visualize_boxes_3d(mirrored.get_box3d_list())
```

```python
# Instantiation Demo
print("\n=== Instantiation Demo ===")
type_list = [Move3D, Union3D, Rect3D, float, float, float, Rect3D, float, float, float, float, float, float]
param_list = [1.0, 1.0, 1.0,  # First Rect3D
              0.5, 2.0, 0.5,   # Second Rect3D
              1.5, 0.0, 0.0]    # Move3D parameters

instantiated_shape = instantiate_3d(type_list, param_list)
print("\nInstantiated shape:")
print(instantiated_shape)
visualize_boxes_3d(instantiated_shape.get_box3d_list())
```
