```python
import torch
from abstractions3d.primitives.shapes import Box3D
from abstractions3d.primitives.visualization import visualize_boxes_3d
from abstractions3d.dsl.nodes import Cube, Move3D, Union3D
from abstractions3d.dsl.instantiation import instantiate_3d
```

```python
# Type list representing the structure of the chair
type_list = [
    Union3D, Union3D, Union3D, Union3D,  # top-level unions
    Cube,                              # seat
    Move3D, Cube, float, float, float, # backrest
    Move3D, Cube, float, float, float, # leg1
    Move3D, Cube, float, float, float, # leg2
    Move3D, Cube, float, float, float, # leg3
    Move3D, Cube, float, float, float, # leg4
    float, float, float                  # params for seat size
]

# Parameters for each shape
param_list = [
    2.0, 0.5, 2.0,           # seat: width, height, depth
    2.0, 2.0, 0.5, 0.0, 1.25, -0.75,  # backrest: w,h,d, x,y,z move
    0.2, 1.0, 0.2, -0.9, -0.75, 0.9,  # leg1
    0.2, 1.0, 0.2, 0.9, -0.75, 0.9,    # leg2
    0.2, 1.0, 0.2, -0.9, -0.75, -0.9,  # leg3
    0.2, 1.0, 0.2, 0.9, -0.75, -0.9    # leg4
]
```

```python
# Build the chair shape from type_list and param_list
chair = instantiate_3d(type_list, param_list)

# Convert to Box3D objects
boxes = chair.get_box_list()

# Visualize the chair in 3D
visualize_boxes_3d(boxes)
```
