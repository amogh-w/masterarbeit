```python
import torch
import random

from abstractions3d.primitives.visualization import visualize_boxes_3d

# Import the shapes
from abstractions3d.data.blueprint import Table3D, Chair3D, Bench3D
```

```python
def sample_uniform(low, high):
    return random.uniform(low, high)
```

```python
table = Table3D(
    top_length=sample_uniform(3.0, 6.0),
    top_depth=sample_uniform(1.5, 3.0),
    top_thickness=sample_uniform(0.1, 0.3),
    leg_length=sample_uniform(0.2, 0.4),
    leg_depth=sample_uniform(0.2, 0.4),
    leg_height=sample_uniform(1.0, 2.0)
)

visualize_boxes_3d(table.get_box3d_list())
```

```python
chair = Chair3D(
    seat_length=sample_uniform(1.5, 2.5),
    seat_depth=sample_uniform(1.5, 2.5),
    seat_thickness=sample_uniform(0.1, 0.3),
    leg_length=sample_uniform(0.15, 0.3),
    leg_depth=sample_uniform(0.15, 0.3),
    leg_height=sample_uniform(0.8, 1.2),
    backrest_height=sample_uniform(1.0, 1.5),
    backrest_thickness=sample_uniform(0.1, 0.3)
)

visualize_boxes_3d(chair.get_box3d_list())
```

```python
bench = Bench3D(
    seat_length=sample_uniform(3.0, 6.0),
    seat_depth=sample_uniform(1.5, 2.5),
    seat_thickness=sample_uniform(0.2, 0.4),
    leg_length=sample_uniform(0.2, 0.4),
    leg_depth=sample_uniform(0.2, 0.4),
    leg_height=sample_uniform(1.0, 1.5),
    backrest_height=sample_uniform(0.8, 1.5),
    backrest_thickness=sample_uniform(0.1, 0.3)
)

visualize_boxes_3d(bench.get_box3d_list())
```
