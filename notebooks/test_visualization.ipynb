```python
import torch
import matplotlib.pyplot as plt

from abstractions.primitives.shapes import Box, Circle
from abstractions.primitives.visualization import show_boxes, show_circles
```

```python
box = Box(center=torch.tensor([0.0, 0.0]), scale=torch.tensor([1.0, 1.0]))
circle = Circle(center=torch.tensor([0.0, 0.0]), radius=torch.tensor(0.5))

# Show box
print("Showing Box:")
show_boxes([box], backend="k3d", lock_2d_view=False)

# Show circle
print("Showing Circle:")
show_circles([circle])
```

```python
# Create 2 boxes
box1 = Box(center=torch.tensor([0.0, 0.0]), scale=torch.tensor([1.0, 1.0]))
box2 = Box(center=torch.tensor([0.5, 0.5]), scale=torch.tensor([0.8, 0.5]))

# Create 2 circles
circle1 = Circle(center=torch.tensor([0.0, 0.0]), radius=torch.tensor(0.5))
circle2 = Circle(center=torch.tensor([0.5, 0.5]), radius=torch.tensor(0.3))

# Show boxes
print("Showing Boxes:")
show_boxes([box1, box2])

# Show circles
print("Showing Circles:")
show_circles([circle1, circle2])
```
