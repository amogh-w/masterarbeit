```python
class SymRef:
    """Mirror symmetry (Reflection) across an axis."""
    def __init__(self, child, axis='AX'):
        self.child = child
        self.axis = axis  # 'AX', 'AY', or 'AZ'

    def __str__(self):
        # Format: (SymRef axis=AX child=(...))
        return f"(SymRef axis={self.axis} child={self.child})"

class SymTrans:
    """Translational symmetry (Repetition)."""
    def __init__(self, child, axis, count, distance):
        self.child = child
        self.axis = axis      # 'AX', 'AY', or 'AZ'
        self.count = count    # Integer: number of repetitions
        self.distance = distance  # Float: spacing between repetitions

    def __str__(self):
        # Format: (SymTrans axis=AX count=3 dist=5.00 child=(...))
        return (f"(SymTrans axis={self.axis} count={self.count} "
                f"dist={self.distance:.2f} child={self.child})")
```

```python
import numpy as np
import copy
import logging
from abstractionsshapecoder.dsl_nodes import Union, Translate, Cuboid, Rotate

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')
logger = logging.getLogger("ShapeOptimizer")

class ShapeOptimizer:
    def __init__(self, threshold=0.01):
        self.threshold = threshold

    def is_near(self, a, b):
        """Helper for scalar comparison (abs:0.01)."""
        return abs(float(a) - float(b)) <= self.threshold

    def is_near_vec(self, v1, v2):
        """Helper for vector comparison."""
        return np.all(np.abs(np.array(v1) - np.array(v2)) <= self.threshold)

    def is_zero(self, val):
        """Rule: zero_add or zero_mov check."""
        if isinstance(val, (list, np.ndarray)):
            return np.all(np.abs(np.array(val)) <= self.threshold)
        return abs(float(val)) <= self.threshold

    def apply_optimization_rules(self, node):
        """
        Implements SEM_3DL_REWRITES logic at the individual node level.
        """
        if node is None: return None

        # --- RULE: s3dl_move_comb & s3dl_zero_mov ---
        if isinstance(node, Translate):
            node.child = self.optimize_node(node.child)
            
            # Combine Move(Move) -> Move(Add(a,x))
            if isinstance(node.child, Translate):
                logger.debug("Rule: s3dl_move_comb applied.")
                new_vec = (np.array(node.vector) + np.array(node.child.vector)).tolist()
                return Translate(node.child.child, new_vec)
            
            # Remove No-op Move
            if self.is_zero(node.vector):
                logger.debug("Rule: s3dl_zero_mov applied.")
                return node.child

        # --- RULE: s3dl_rot_comb & s3dl_zero_rot ---
        if isinstance(node, Rotate):
            node.child = self.optimize_node(node.child)
            
            # Combine Rotate(Rotate) if axis is same
            if isinstance(node.child, Rotate) and node.axis == node.child.axis:
                logger.debug("Rule: s3dl_rot_comb applied.")
                return Rotate(node.child.child, node.axis, node.angle + node.child.angle)
            
            # Remove No-op Rotate
            if self.is_zero(node.angle):
                logger.debug("Rule: s3dl_zero_rot applied.")
                return node.child

        # --- RULE: s3dl_rot_over_move ---
        # (Move (Rotate X A v) a b c) -> (Rotate (Move X a b c) A v)
        if isinstance(node, Translate) and isinstance(node.child, Rotate):
            logger.debug("Rule: s3dl_rot_over_move applied.")
            rot_node = node.child
            inner_move = Translate(rot_node.child, node.vector)
            return Rotate(inner_move, rot_node.axis, rot_node.angle)

        # --- RULE: s3dl_union_over_sym ---
        # (SymRef (Union X Y) A) -> (Union (SymRef X A) (SymRef Y A))
        if isinstance(node, SymRef) and isinstance(node.child, Union):
            logger.debug("Rule: s3dl_union_over_sym applied.")
            left_sym = SymRef(node.child.left, node.axis)
            right_sym = SymRef(node.child.right, node.axis)
            return Union(left_sym, right_sym)

        return node

    def find_symmetries(self, parts):
        optimized = []
        used_indices = set()

        for i in range(len(parts)):
            if i in used_indices: continue
            p1 = parts[i]
            
            # --- 1. Translational Symmetry (SymTrans) ---
            # Look for 3+ items with equal spacing along an axis
            for axis_idx, axis_name in enumerate(['AX', 'AY', 'AZ']):
                sequence = [i]
                if not isinstance(p1, Translate): continue
                
                # Find all potential matches in this axis
                for j in range(i + 1, len(parts)):
                    if j in used_indices: continue
                    p2 = parts[j]
                    if isinstance(p2, Translate) and self.is_near_node_geometry(p1.child, p2.child):
                        # Check if it aligns on the current axis but matches others
                        diff = np.array(p1.vector) - np.array(p2.vector)
                        other_axes = [0, 1, 2]
                        other_axes.remove(axis_idx)
                        
                        # Must be aligned on other axes
                        if all(abs(diff[ax]) < self.threshold for ax in other_axes):
                            sequence.append(j)
                
                if len(sequence) >= 3:
                    # Sort by the active axis to find consistent distance
                    sequence.sort(key=lambda idx: parts[idx].vector[axis_idx])
                    vecs = [np.array(parts[idx].vector)[axis_idx] for idx in sequence]
                    dist = vecs[1] - vecs[0]
                    
                    # Verify spacing is consistent
                    is_consistent = all(self.is_near(vecs[k+1] - vecs[k], dist) for k in range(len(vecs)-1))
                    
                    if is_consistent:
                        logger.info(f"Rule: SymTrans detected ({len(sequence)} parts, axis {axis_name})")
                        # Representative is the first item
                        optimized.append(SymTrans(parts[sequence[0]], axis_name, len(sequence), dist))
                        used_indices.update(sequence)
                        break

            if i in used_indices: continue

            # --- 2. Mirror Symmetry (SymRef) ---
            found_mirror = False
            for j in range(i + 1, len(parts)):
                if j in used_indices: continue
                p2 = parts[j]
                
                if isinstance(p1, Translate) and isinstance(p2, Translate):
                    v1, v2 = np.array(p1.vector), np.array(p2.vector)
                    for axis_name, idx in [('AX', 0), ('AY', 1), ('AZ', 2)]:
                        mirrored_v2 = v2.copy()
                        mirrored_v2[idx] = -mirrored_v2[idx]
                        
                        if self.is_near_vec(v1, mirrored_v2) and self.is_near_node_geometry(p1.child, p2.child):
                            optimized.append(SymRef(p1, axis_name))
                            used_indices.add(i); used_indices.add(j)
                            found_mirror = True
                            break
                    if found_mirror: break
            
            if not found_mirror:
                optimized.append(p1)
                used_indices.add(i)
                
        return optimized

    def is_near_node_geometry(self, n1, n2):
        """Checks if two nodes are identical in shape/size (for symmetry)."""
        if type(n1) != type(n2): return False
        if isinstance(n1, Cuboid):
            return self.is_near_vec(n1.size, n2.size)
        # Add more geometry checks (Sphere, Cylinder) here
        return False

    def flatten_unions(self, node):
        if isinstance(node, Union):
            return self.flatten_unions(node.left) + self.flatten_unions(node.right)
        return [node] if node is not None else []

    def rebuild_union(self, nodes):
        if not nodes: return None
        if len(nodes) == 1: return nodes[0]
        # s3dl_union_over ensures we build a consistent tree
        return Union(nodes[0], self.rebuild_union(nodes[1:]))

    def optimize_node(self, node):
        return self.apply_optimization_rules(node)

    def run(self, root_node):
        if root_node is None: return None

        # 1. Deep apply local rules (combining moves, rotations, etc.)
        root_node = self.optimize_node(root_node)
        
        # 2. Flatten for global structural rules
        parts = self.flatten_unions(root_node)
        
        # 3. Apply Symmetry and Factoring
        parts = self.find_symmetries(parts)
        
        # 4. Final rebuild
        return self.rebuild_union(parts)

# --- EXECUTION ---
optimizer = ShapeOptimizer()
results = {}

# Example loop with improved error handling
for shape_id, data in shapes_l0.items():
    try:
        original_dsl = data.get("dsl")
        if original_dsl is None:
            continue

        optimized = optimizer.run(copy.deepcopy(original_dsl))
        results[shape_id] = optimized
        
        orig_count = len(optimizer.flatten_unions(original_dsl))
        opt_count = len(optimizer.flatten_unions(optimized))
        
        if opt_count < orig_count:
            print(f"✅ {shape_id}: Reduced from {orig_count} to {opt_count} nodes.")
            
    except Exception as e:
        logger.error(f"Critical failure on {shape_id}: {e}", exc_info=True)
```

```python
import pandas as pd
import copy
import logging

# Helper to turn the node tree into a string for the CSV
def dsl_to_string(node):
    if node is None:
        return ""
    # This assumes your DSL nodes have a __repr__ or __str__ 
    # If not, we can use: str(type(node).__name__)
    return str(node)

stats_list = []

for shape_id, data in shapes_l0.items():
    try:
        original_dsl = data.get("dsl")
        if original_dsl is None:
            logger.warning(f"SKIP: {shape_id} - No DSL found.")
            continue

        # 1. Run Optimization
        # We keep a copy of the original for the 'before' string
        original_copy = copy.deepcopy(original_dsl)
        optimized = optimizer.run(copy.deepcopy(original_dsl))
        
        # 2. Extract Metrics
        orig_count = len(optimizer.flatten_unions(original_copy))
        opt_count = len(optimizer.flatten_unions(optimized))
        reduction = orig_count - opt_count
        
        # 3. Logging (No emojis)
        logger.info(f"Shape: {shape_id} | Reduction: {reduction}")
        
        # 4. Append to list for DataFrame
        stats_list.append({
            "shape_id": shape_id,
            "status": "SUCCESS",
            "original_count": orig_count,
            "optimized_count": opt_count,
            "reduction": reduction,
            "original_dsl": dsl_to_string(original_copy),
            "optimized_dsl": dsl_to_string(optimized)
        })
            
    except Exception as e:
        logger.error(f"FAILURE: {shape_id} - {str(e)}")
        stats_list.append({
            "shape_id": shape_id,
            "status": f"ERROR: {str(e)}",
            "original_count": None,
            "optimized_count": None,
            "reduction": None,
            "original_dsl": None,
            "optimized_dsl": None
        })

# --- Pandas Processing ---
df = pd.DataFrame(stats_list)

# Calculate percentage
df['reduction_pct'] = (df['reduction'] / df['original_count'] * 100).round(2).fillna(0)

# Reorder columns for better readability in CSV
column_order = [
    "shape_id", "status", "original_count", "optimized_count", 
    "reduction", "reduction_pct", "original_dsl", "optimized_dsl"
]
df = df[column_order]

# Save to CSV
df.to_csv("shape_optimization_report.csv", index=False)

logger.info("CSV Report generated with 'original_dsl' and 'optimized_dsl' columns.")
```

```python
from __future__ import annotations
from egglog import *

egraph = EGraph()

# 1. Define your Sorts (Types)
class Shape(Expr):
    def __init__(self): ...
    
    @classmethod
    def cuboid(cls, w: f64, h: f64, d: f64) -> Shape: ...
    
    def translate(self, x: f64, y: f64, z: f64) -> Shape: ...
    
    def union(self, other: Shape) -> Shape: ...
    
    def sym_ref(self, axis: String) -> Shape: ...

# 2. Define the Rewrite Rules
@egraph.register
def rules(s: Shape, x1: f64, y1: f64, z1: f64, x2: f64, y2: f64, z2: f64, axis: String):
    # s3dl_move_comb: Combine nested translations
    yield rewrite(s.translate(x1, y1, z1).translate(x2, y2, z2)).to(
        s.translate(x1 + x2, y1 + y2, z1 + z2)
    )
    
    # s3dl_zero_mov: Remove identity moves
    yield rewrite(s.translate(0.0, 0.0, 0.0)).to(s)
    
    # s3dl_sym_x_def: Recognize X-Symmetry
    # If we see a union of a shape and its mirrored version, replace with SymRef
    yield rewrite(
        s.translate(x1, y1, z1).union(s.translate(-x1, y1, z1))
    ).to(
        s.translate(x1, y1, z1).sym_ref("AX")
    )

# 3. Usage
# Create a messy expression
original = Shape.cuboid(1.0, 1.0, 1.0).translate(5.0, 0.0, 0.0).translate(2.0, 0.0, 0.0)
expr = egraph.let("my_shape", original)

# Run the rules until saturation
egraph.run(10)

# Extract the best (simplest) version
optimized = egraph.extract(expr)
print(f"Optimized DSL: {optimized}")
```

```python
from __future__ import annotations
from egglog import *

def optimize_shape_collection():
    egraph = EGraph()

    # 1. DEFINE THE SORT AND FUNCTIONS
    class Shape(Expr):
        def __init__(self) -> None: ...

    # We define these as functions that take Shape as the first argument
    @function
    def cuboid(w: f64, h: f64, d: f64) -> Shape: ...

    @function
    def translate(s: Shape, x: f64, y: f64, z: f64) -> Shape: ...

    @function
    def union(s1: Shape, s2: Shape) -> Shape: ...

    @function
    def sym_ref(s: Shape, axis: String) -> Shape: ...

    # 2. DEFINE THE OPTIMIZATION RULES
    @egraph.register
    def rules(s: Shape, s2: Shape, s3: Shape,
              x1: f64, y1: f64, z1: f64, 
              x2: f64, y2: f64, z2: f64):
        
        # Combine nested translations
        yield rewrite(translate(translate(s, x1, y1, z1), x2, y2, z2)).to(
            translate(s, x1 + x2, y1 + y2, z1 + z2)
        )
        
        # Remove identity translations
        yield rewrite(translate(s, 0.0, 0.0, 0.0)).to(s)

        # Union Commutativity
        yield rewrite(union(s, s2)).to(union(s2, s))
        
        # Factoring common moves
        yield rewrite(
            union(translate(s, x1, y1, z1), translate(s2, x1, y1, z1))
        ).to(
            translate(union(s, s2), x1, y1, z1)
        )

        # Symmetry Detection (X-Axis)
        yield rewrite(
            union(translate(s, x1, y1, z1), translate(s, -x1, y1, z1))
        ).to(
            sym_ref(translate(s, x1, y1, z1), "AX")
        )

    # 3. CONSTRUCT A TEST CASE
    box = cuboid(1.0, 1.0, 1.0)
    
    # Left side: -5.0
    # Right side: 3.0 + 2.0 = 5.0
    part_left = translate(box, -5.0, 0.0, 0.0)
    part_right = translate(translate(box, 3.0, 0.0, 0.0), 2.0, 0.0, 0.0)
    
    complex_shape = union(part_left, part_right)
    
    # 4. RUN AND EXTRACT
    expr = egraph.let("my_shape", complex_shape)
    
    print("Running Equality Saturation...")
    egraph.run(10)
    
    optimized = egraph.extract(expr)
    
    print("-" * 30)
    print(f"Original: {complex_shape}")
    print("-" * 30)
    print(f"Optimized: {optimized}")
    print("-" * 30)

if __name__ == "__main__":
    optimize_shape_collection()
```

```python
from __future__ import annotations
from egglog import *

egraph = EGraph()

# 1. Define the Sort
class Shape(Expr):
    def __init__(self) -> None: ...

# 2. Define the Functions (Constructors)
@function
def cuboid(w: f64, h: f64, d: f64) -> Shape: ...

@function
def translate(s: Shape, x: f64, y: f64, z: f64) -> Shape: ...

@function
def union(s1: Shape, s2: Shape) -> Shape: ...

@function
def sym_ref(s: Shape, axis: String) -> Shape: ...

@function
def sym_trans(s: Shape, axis: String, count: f64, dist: f64) -> Shape: ...

print("Sorts and Functions defined.")
```

```python
@egraph.register
def optimization_rules(s: Shape, s2: Shape, s3: Shape,
                       x1: f64, y1: f64, z1: f64, 
                       x2: f64, y2: f64, z2: f64,
                       axis: String):
    
    # Rule 1: Combine nested translations
    yield rewrite(translate(translate(s, x1, y1, z1), x2, y2, z2)).to(
        translate(s, x1 + x2, y1 + y2, z1 + z2)
    )
    
    # Rule 2: Remove identity (zero) translations
    yield rewrite(translate(s, 0.0, 0.0, 0.0)).to(s)

    # Rule 3: Union Commutativity (A+B = B+A)
    # This is critical so the engine can find symmetries regardless of order
    yield rewrite(union(s, s2)).to(union(s2, s))
    
    # Rule 4: Union Associativity
    yield rewrite(union(s, union(s2, s3))).to(union(union(s, s2), s3))
    
    # Rule 5: Factoring common moves
    # Union(Move(A, v), Move(B, v)) -> Move(Union(A, B), v)
    yield rewrite(
        union(translate(s, x1, y1, z1), translate(s2, x1, y1, z1))
    ).to(
        translate(union(s, s2), x1, y1, z1)
    )

    # Rule 6: X-Axis Symmetry Detection
    yield rewrite(
        union(translate(s, x1, y1, z1), translate(s, -x1, y1, z1))
    ).to(
        sym_ref(translate(s, x1, y1, z1), "AX")
    )

print("Optimization rules registered.")
```

```python
def to_egglog(node):
    """
    Recursively converts your DSL nodes into egglog expressions.
    This assumes your nodes have .child, .left, .right, .vector, or .size attributes.
    """
    # Import inside the function to ensure access to your node types
    from abstractionsshapecoder.dsl_nodes import Union, Translate, Cuboid
    
    if isinstance(node, Cuboid):
        return cuboid(float(node.size[0]), float(node.size[1]), float(node.size[2]))
    
    if isinstance(node, Translate):
        inner = to_egglog(node.child)
        v = node.vector
        return translate(inner, float(v[0]), float(v[1]), float(v[2]))
    
    if isinstance(node, Union):
        l = to_egglog(node.left)
        r = to_egglog(node.right)
        return union(l, r)
    
    # If it's already an egglog expression or a type we don't recognize
    return node

print("Translator defined.")
```

```python
import re
from abstractionsshapecoder.dsl_nodes import Union, Translate, Cuboid

def from_egglog(expr):
    """
    Converts an extracted egglog expression by parsing its string representation.
    """
    # Clean up the string to remove module prefixes like '__main___'
    s = str(expr).replace("__main___", "")

    # Define the environment for eval to map egglog names to your classes
    # We create helper functions that match the egglog function signatures
    env = {
        "cuboid": lambda w, h, d: Cuboid([w, h, d]),
        "translate": lambda child, x, y, z: Translate(child, [x, y, z]),
        "union": lambda l, r: Union(l, r),
        "sym_ref": lambda child, axis: SymRef(child, axis),
        "sym_trans": lambda child, axis, count, dist: SymTrans(child, axis, int(count), dist)
    }

    try:
        # eval() here is safe because it only uses the controlled 'env' 
        # and the string generated by egglog.
        return eval(s, {"__builtins__": {}}, env)
    except Exception as e:
        print(f"Error parsing egglog string: {e}")
        print(f"Original string was: {s}")
        return None

print("Back-Translator defined.")
```

```python
import time
from abstractionsshapecoder.dsl_nodes import Union, Translate, Cuboid

# 1. Create messy DSL
box = Cuboid([1, 1, 1])
part_a = Translate(Translate(box, [3, 0, 0]), [2, 0, 0])
part_b = Translate(box, [-5, 0, 0])
test_shape = Union(part_a, part_b)

# 2. Run Pipeline
unique_id = int(time.time() * 1000) # millisecond precision
egg_expr = to_egglog(test_shape)
handle = egraph.let(f"test_run_{unique_id}", egg_expr)
egraph.run(10)
best_egg = egraph.extract(handle)

# 3. Convert back
optimized_dsl = from_egglog(best_egg)

print("--- RESULTS ---")
print(f"Original  : {test_shape}")
print(f"Optimized : {optimized_dsl}")
```
