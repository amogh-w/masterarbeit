```python
import pickle
from pathlib import Path
from abstractionsshapecoder.debug_utils import debug_info, debug_success, debug_error

def load_processed_datasets():
    """
    Loads all hierarchical datasets (L0, L1-VAE, L1-AE, L2-AE) from the saved directory.
    """
    SAVED_DIR = Path.cwd().parent / "src" / "abstractionsshapecoder" / "saved"
    
    # Define file map: {Variable_Name: Filename}
    dataset_files = {
        "shapes_l0": "all_dsl_shapes.pkl",
        "shapes_l1_vae": "all_abstracted_shapes_L1_VAE.pkl",
        "shapes_l1_ae": "all_abstracted_shapes_L1_AE.pkl",
        "shapes_l2_ae": "all_abstracted_shapes_L2_AE.pkl"
    }
    
    loaded_datasets = {}
    
    for var_name, filename in dataset_files.items():
        file_path = SAVED_DIR / filename
        if file_path.exists():
            debug_info(f"Loading {var_name} from {filename}...")
            with open(file_path, "rb") as f:
                loaded_datasets[var_name] = pickle.load(f)
            debug_success(f"Successfully loaded {len(loaded_datasets[var_name])} shapes into {var_name}.")
        else:
            debug_error(f"Missing file: {filename}. Ensure previous training cells were run.")
            loaded_datasets[var_name] = {}
            
    return loaded_datasets

# Execute the load
datasets = load_processed_datasets()

# Extract into global namespace for easier access in subsequent cells
shapes_l0 = datasets.get("shapes_l0")
```

```python
class SymRef:
    """Mirror symmetry (Reflection) across an axis."""
    def __init__(self, child, axis='AX'):
        self.child = child
        self.axis = axis  # 'AX', 'AY', or 'AZ'

    def __str__(self):
        # Format: (SymRef axis=AX child=(...))
        return f"(SymRef axis={self.axis} child={self.child})"

class SymTrans:
    """Translational symmetry (Repetition)."""
    def __init__(self, child, axis, count, distance):
        self.child = child
        self.axis = axis      # 'AX', 'AY', or 'AZ'
        self.count = count    # Integer: number of repetitions
        self.distance = distance  # Float: spacing between repetitions

    def __str__(self):
        # Format: (SymTrans axis=AX count=3 dist=5.00 child=(...))
        return (f"(SymTrans axis={self.axis} count={self.count} "
                f"dist={self.distance:.2f} child={self.child})")
```

```python
import pandas as pd
import time
from tqdm import tqdm
from egglog import *
from abstractionsshapecoder.dsl_nodes import Union, Translate, Cuboid, Rotate

# --- 1. INITIALIZE E-GRAPH ---
egraph = EGraph()
EPSILON = 0.64

class Shape(Expr):
    def __init__(self) -> None: ...

# --- 2. DEFINE E-GRAPH FUNCTIONS WITH COSTS ---
@function(cost=10)
def cuboid(w: f64, h: f64, d: f64) -> Shape: ...

@function(cost=10)
def translate(s: Shape, x: f64, y: f64, z: f64) -> Shape: ...

@function(cost=10)
def rotate(s: Shape, q0: f64, q1: f64, q2: f64, q3: f64) -> Shape: ...

@function(cost=20) # High cost for unions to encourage elimination
def union_op(s1: Shape, s2: Shape) -> Shape: ...

@function(cost=5) # Low cost for symmetry (preferred)
def sym_ref(s: Shape, axis: String) -> Shape: ...

# --- 3. REGISTER GUARDED OPTIMIZATION RULES ---
@egraph.register
def optimization_rules(s: Shape, s2: Shape, s3: Shape,
                       x: f64, y: f64, z: f64, 
                       x2: f64, y2: f64, z2: f64,
                       q0: f64, q1: f64, q2: f64, q3: f64,
                       axis: String):
    
    # 1. Identity Translation (with tolerance)
    yield rule(
        eq(s2).to(translate(s, x, y, z)),
        (x * x < EPSILON), (y * y < EPSILON), (z * z < EPSILON)
    ).then(
        union(s2).with_(s)
    )

    # 2. Nesting: translate(translate(s)) -> translate(sum)
    yield rewrite(translate(translate(s, x, y, z), x2, y2, z2)).to(
        translate(s, x + x2, y + y2, z + z2)
    )

    # 3. Factoring (The "Compressor") with tolerance
    yield rule(
        eq(s3).to(union_op(translate(s, x, y, z), translate(s2, x2, y2, z2))),
        ((x - x2) * (x - x2) < EPSILON),
        ((y - y2) * (y - y2) < EPSILON),
        ((z - z2) * (z - z2) < EPSILON)
    ).then(
        union(s3).with_(translate(union_op(s, s2), x, y, z))
    )

    # 4. Symmetry Detection with Tolerance
    # X-Axis
    yield rule(
        eq(s3).to(union_op(translate(s, x, y, z), translate(s, x2, y2, z2))),
        ((y - y2) * (y - y2) < EPSILON),
        ((z - z2) * (z - z2) < EPSILON),
        ((x + x2) * (x + x2) < EPSILON)
    ).then(
        union(s3).with_(sym_ref(translate(s, x, y, z), "AX"))
    )
    
    # Y-Axis
    yield rule(
        eq(s3).to(union_op(translate(s, x, y, z), translate(s, x2, y2, z2))),
        ((x - x2) * (x - x2) < EPSILON),
        ((z - z2) * (z - z2) < EPSILON),
        ((y + y2) * (y + y2) < EPSILON)
    ).then(
        union(s3).with_(sym_ref(translate(s, x, y, z), "AY"))
    )

    # 5. Consolidation & Rotation Pushing
    yield rewrite(union_op(sym_ref(s, axis), sym_ref(s2, axis))).to(
        sym_ref(union_op(s, s2), axis)
    )

    yield rewrite(rotate(union_op(s, s2), q0, q1, q2, q3)).to(
        union_op(rotate(s, q0, q1, q2, q3), rotate(s2, q0, q1, q2, q3))
    )

# --- 4. TRANSLATION HELPERS ---
def count_parts(node):
    if isinstance(node, Union):
        return count_parts(node.left) + count_parts(node.right)
    if hasattr(node, 'child'): # For Translate/Rotate
        return count_parts(node.child)
    return 1

def to_egglog(node):
    if isinstance(node, Cuboid):
        return cuboid(float(node.size[0]), float(node.size[1]), float(node.size[2]))
    if isinstance(node, Translate):
        return translate(to_egglog(node.child), float(node.vector[0]), float(node.vector[1]), float(node.vector[2]))
    if isinstance(node, Rotate):
        q = node.quaternion 
        return rotate(to_egglog(node.child), float(q[0]), float(q[1]), float(q[2]), float(q[3]))
    if isinstance(node, Union):
        return union_op(to_egglog(node.left), to_egglog(node.right))
    return node

def from_egglog(expr):
    # Ensure SymRef handles correctly in your env
    s = str(expr).replace("__main__.", "")
    env = {
        "cuboid": lambda w, h, d: Cuboid([w, h, d]),
        "translate": lambda child, x, y, z: Translate(child, [x, y, z]),
        "rotate": lambda child, q0, q1, q2, q3: Rotate(child, [q0, q1, q2, q3]),
        "union_op": lambda l, r: Union(l, r),
        "sym_ref": lambda child, axis: f"SymRef({child}, {axis})", # Replace with actual SymRef constructor if imported
    }
    try:
        return eval(s, {"__builtins__": {}}, env)
    except Exception as e:
        raise ValueError(f"Eval Error: {e} on string: {s}")

# --- 5. BATCH EXECUTION LOOP ---
stats = []
results = {}

# Assuming shapes_l0 is your input dictionary
print("Starting batch optimization with egglog (Guarded Rules)...")
items = list(shapes_l0.items())
items = items[:100]

for i, (shape_id, data) in enumerate(tqdm(items, desc="Optimizing Shapes")):
    try:
        orig_node = data.get("dsl")
        if orig_node is None: continue
            
        egraph.push() 
        egg_expr = to_egglog(orig_node)
        handle = egraph.let(f"sh_{i}", egg_expr)
        
        # Run saturation
        egraph.run(4)
        
        best_egg = egraph.extract(handle)
        optimized_node = from_egglog(best_egg)
        egraph.pop()
        
        orig_count = count_parts(orig_node)
        # Handle cases where optimized_node might be a string (from sym_ref eval) or a DSL node
        opt_count = count_parts(optimized_node) if not isinstance(optimized_node, str) else 1 
        
        results[shape_id] = optimized_node
        stats.append({
            "shape_id": shape_id,
            "status": "SUCCESS",
            "original_count": orig_count,
            "optimized_count": opt_count,
            "reduction": orig_count - opt_count,
            "reduction_pct": f"{((orig_count - opt_count)/orig_count*100):.2f}%" if orig_count > 0 else "0%",
        })

    except Exception as e:
        stats.append({"shape_id": shape_id, "status": f"ERROR: {str(e)[:50]}"})

# --- 6. SAVE ---
df = pd.DataFrame(stats)
df.to_csv("egglog_optimization_report.csv", index=False)
print("\nProcessing Complete!")
print(f"Average Reduction: {df.loc[df['status']=='SUCCESS', 'reduction'].mean():.2f} nodes")
```

```python
import numpy as np
from abstractionsshapecoder.dsl_nodes import Union, Translate

def get_all_translations(node):
    """Recursively finds all translation vectors in the DSL tree."""
    vectors = []
    if isinstance(node, Translate):
        vectors.append(np.array(node.vector, dtype=float))
        vectors.extend(get_all_translations(node.child))
    elif isinstance(node, Union):
        vectors.extend(get_all_translations(node.left))
        vectors.extend(get_all_translations(node.right))
    elif hasattr(node, 'child'):
        vectors.extend(get_all_translations(node.child))
    return vectors

def analyze_noise(shapes_dict):
    all_diffs = []
    
    for shape_id, data in shapes_dict.items():
        node = data.get("dsl")
        if not node: continue
        
        vecs = get_all_translations(node)
        if len(vecs) < 2: continue
        
        # Compare pairs to find the "jitter" between objects
        for i in range(len(vecs)):
            for j in range(i + 1, len(vecs)):
                # We care about differences that are almost zero or almost symmetric
                diff = np.abs(vecs[i] - vecs[j])
                sum_val = np.abs(vecs[i] + vecs[j])
                
                # Capture the smallest non-zero differences
                for val in list(diff) + list(sum_val):
                    if 1e-9 < val < 0.5: # Ignore exact matches and large gaps
                        all_diffs.append(val)

    if not all_diffs:
        print("No jitter detected. Your data might already be perfectly integer-aligned.")
        return 0.001

    all_diffs = np.array(all_diffs)
    print(f"--- Noise Analysis ---")
    print(f"Min Jitter:  {np.min(all_diffs):.6f}")
    print(f"Max Jitter:  {np.max(all_diffs):.6f}")
    print(f"Mean Jitter: {np.mean(all_diffs):.6f}")
    print(f"90th Percentile: {np.percentile(all_diffs, 90):.6f}")
    
    # The 'best' epsilon is usually slightly above the 90th percentile of noise
    suggested_eps = np.percentile(all_diffs, 90) * 1.5
    return suggested_eps

# Run it
best_eps = analyze_noise(shapes_l0)
print(f"\nRecommended EPSILON for your code: {best_eps:.6f}")
```
