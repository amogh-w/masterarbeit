# Bookshelf 1

```python
from abstractions.data.bookshelf_1 import bookshelf_1

# Define sliders for bookshelf parameters
width_slider = widgets.FloatSlider(value=1.0, min=0.5, max=1.5, step=0.05, description="Width:")
height_slider = widgets.FloatSlider(value=2.0, min=1.0, max=3.0, step=0.05, description="Height:")
shelf_thickness_slider = widgets.FloatSlider(value=0.05, min=0.02, max=0.15, step=0.01, description="Shelf Thick:")
num_shelves_slider = widgets.IntSlider(value=4, min=1, max=8, description="Num Shelves:")
num_dividers_slider = widgets.IntSlider(value=2, min=0, max=6, description="Num Dividers:")
divider_thickness_slider = widgets.FloatSlider(value=0.05, min=0.02, max=0.15, step=0.01, description="Divider Thick:")

output = widgets.Output()

def update_bookshelf(change=None):
    output.clear_output()
    with output:
        shape = bookshelf_1(
            width=width_slider.value,
            height=height_slider.value,
            shelf_thickness=shelf_thickness_slider.value,
            num_shelves=num_shelves_slider.value,
            num_dividers=num_dividers_slider.value,
            divider_thickness=divider_thickness_slider.value,
        )
        print_tree(shape)
        show_boxes(shape.get_box_list(), backend="plotly")

# Attach updates to sliders
for slider in [
    width_slider,
    height_slider,
    shelf_thickness_slider,
    num_shelves_slider,
    num_dividers_slider,
    divider_thickness_slider,
]:
    slider.observe(update_bookshelf, names="value")

# Display widget and output area
display(
    widgets.VBox([
        width_slider,
        height_slider,
        shelf_thickness_slider,
        num_shelves_slider,
        num_dividers_slider,
        divider_thickness_slider,
        output,
    ])
)

# Initialize the visualization
update_bookshelf()
```

# Chair 1

```python
from abstractions.data.chair_1 import chair_1

# Define interactive sliders
width = widgets.FloatSlider(value=0.8, min=0.3, max=1.5, step=0.05, description="Width:")
leg_height = widgets.FloatSlider(value=0.6, min=0.2, max=1.5, step=0.05, description="Leg Height:")
leg_thickness = widgets.FloatSlider(value=0.1, min=0.05, max=0.3, step=0.01, description="Leg Thick:")
back_height = widgets.FloatSlider(value=0.8, min=0.3, max=1.5, step=0.05, description="Back Height:")
back_thickness = widgets.FloatSlider(value=0.1, min=0.05, max=0.3, step=0.01, description="Back Thick:")
seat_thickness = widgets.FloatSlider(value=0.1, min=0.05, max=0.3, step=0.01, description="Seat Thick:")

# Output area
output = widgets.Output()

def update(change=None):
    output.clear_output()
    with output:
        shape = chair_1(
            width=width.value,
            leg_height=leg_height.value,
            leg_thickness=leg_thickness.value,
            back_height=back_height.value,
            back_thickness=back_thickness.value,
            seat_thickness=seat_thickness.value,
        )
        print_tree(shape)
        show_boxes(shape.get_box_list(), backend="plotly")

# Link sliders to update
for slider in [width, leg_height, leg_thickness, back_height, back_thickness, seat_thickness]:
    slider.observe(update, names='value')

# Display everything
display(widgets.VBox([
    width, leg_height, leg_thickness,
    back_height, back_thickness, seat_thickness,
    output,
]))

update()
```

# Chair 3

```python
from abstractions.data.chair_3 import chair_3

# Create sliders for all parameters
width_slider = widgets.FloatSlider(value=0.75, min=0.5, max=1.0, step=0.025, description='Width:')
leg_height_slider = widgets.FloatSlider(value=0.75, min=0.5, max=1.0, step=0.025, description='Leg Height:')
leg_thickness_slider = widgets.FloatSlider(value=0.1, min=0.05, max=0.2, step=0.01, description='Leg Thickness:')
back_height_slider = widgets.FloatSlider(value=0.6, min=0.2, max=1.0, step=0.025, description='Back Height:')
back_thickness_slider = widgets.FloatSlider(value=0.1, min=0.05, max=0.2, step=0.01, description='Back Thickness:')
seat_thickness_slider = widgets.FloatSlider(value=0.1, min=0.05, max=0.2, step=0.01, description='Seat Thickness:')
armrest_thickness_slider = widgets.FloatSlider(value=0.1, min=0.05, max=0.15, step=0.01, description='Armrest Thickness:')

output = widgets.Output()

def update_chair(_):
    output.clear_output()
    with output:
        shape = chair_3(
            width=width_slider.value,
            leg_height=leg_height_slider.value,
            leg_thickness=leg_thickness_slider.value,
            back_height=back_height_slider.value,
            back_thickness=back_thickness_slider.value,
            seat_thickness=seat_thickness_slider.value,
            armrest_thickness=armrest_thickness_slider.value,
        )
        display(Markdown("### DSL Tree"))
        print_tree(shape)
        display(Markdown("### Geometry"))
        show_boxes(shape.get_box_list(), backend="plotly")

# Link sliders to update function
for slider in [width_slider, leg_height_slider, leg_thickness_slider, back_height_slider,
               back_thickness_slider, seat_thickness_slider, armrest_thickness_slider]:
    slider.observe(update_chair, names='value')

# Display all widgets and output
ui = widgets.VBox([
    width_slider, leg_height_slider, leg_thickness_slider, back_height_slider,
    back_thickness_slider, seat_thickness_slider, armrest_thickness_slider
])

display(ui, output)

# Trigger initial update
update_chair(None)
```
