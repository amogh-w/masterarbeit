```python
# Notebook: view_chair.ipynb

# 1. Imports
import torch
from dataloader_symh import Data_Loader
from draw3dobb import showGenshape, tryPlot

# 2. Set dataset path
dataset_dir = './dropbox/Chair/'  # adjust if needed

# 3. Load chair dataset (example: first 10 shapes)
split_num = 0
total_num = 10
train_loader = Data_Loader(dataset_dir, is_train=True, split_num=split_num, total_num=total_num)

# 4. Select a chair to visualize
chair_tree = train_loader[0]  # first chair

# 5. Visualize the bounding boxes for this chair
# Recover the bounding boxes from the tree
recover_boxes = []
for node in chair_tree.trees if hasattr(chair_tree, 'trees') else [chair_tree]:
    try:
        for child in node.root.left, node.root.right:
            if hasattr(child, 'leaf_points'):
                recover_boxes.append(child.leaf_points.squeeze(0).numpy())
    except:
        continue

# Use the helper from draw3dobb.py
showGenshapes([recover_boxes])
```

```python
# Notebook: view_chair_boxes.ipynb

import os
import numpy as np
from scipy.io import loadmat
from draw3dobb import tryPlot, draw

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Set path to Chair boxes
chair_boxes_dir = './dropbox/Chair/boxes/'

# 2. List a few box files
box_files = sorted(os.listdir(chair_boxes_dir))
print("Available box files:", box_files[:5])

# 3. Load one box
box_path = os.path.join(chair_boxes_dir, box_files[0])
box_data = loadmat(box_path)['box']  # adjust key if needed

# 4. Plot each box
fig = plt.figure()
ax = Axes3D(fig)
cmap = plt.get_cmap('jet_r')

for i, b in enumerate(box_data):
    draw(ax, b.flatten(), cmap(float(i)/len(box_data)))

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
